```python
import numpy as np
import pandas as pd
import pylab as plt
```

```python
x = np.arange(0,10,0.1)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.rc('font',family='Microsoft yahei')
plt.plot(x,y1,label = 'y1')
plt.plot(x,y2,label = 'y2')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.title('图1')
plt.legend()
```

```python
plt.plot(x,y1, marker='o', markerfacecolor='white')
plt.plot(x,y2, marker='o', markerfacecolor='white')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.title('图3')
```

```python
plt.scatter(x,y1,marker='o',c = 'white',edgecolors='#1f77b4')
plt.scatter(x,y2,marker = 'o',c = 'white',edgecolors='#ff7f0e')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.title('图2')
```
